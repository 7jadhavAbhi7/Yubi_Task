```python
import ipywidgets as widgets
import glob
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

# Get the list of images from the 'examples/source_image' folder
print("Choose the image name to animate:")
img_list = glob.glob('examples/source_image/*.png')
img_list = sorted([img.split('/')[-1].split('.')[0] for img in img_list])  # Extract filenames without extensions

# Create a dropdown widget with no default selection
default_head_name = widgets.Dropdown(options=img_list, description="Image:", value=None)

# Output widget to display the image
output = widgets.Output()

# Function to update the displayed image
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value' and change['new']:
        with output:
            clear_output(wait=True)  # Clear previous output
            plt.figure(figsize=(5, 5))
            img_path = f'examples/source_image/{change["new"]}.png'
            plt.imshow(plt.imread(img_path))
            plt.axis('off')
            plt.show()

# Attach the observer function to the dropdown
default_head_name.observe(on_change, names='value')

# Display the dropdown and output widget
display(default_head_name, output)
```
